```python
import numpy as np
import pandas as pd
```

```python
bike_df=pd.read_csv('Used_Bikes.csv')
```

```python
bike_df.shape
```

```python
bike_df=bike_df[bike_df['price']<500000]
```

```python
bike_df.shape
```

```python
bike_df.head()
```

```python
import re
```

```python
def bike_name_cleaner(obj):
    pattern = r'[0-9]'
    mod_string = re.sub(pattern, '', obj)
    mod_string= mod_string.replace('cc','')
    return mod_string
  
```

```python
bike_df['bike_name']=bike_df['bike_name'].apply(bike_name_cleaner)
```

```python
bike_df.head(20)
```

```python
bike_df=bike_df.drop(columns=['owner'])
```

```python
bike_df.head()
```

### split

split the data transfrom the data( encode coloumn) apply model make pipeline and fit it dump pipeline

```python
X=bike_df.drop(columns=['price'])
```

```python
X.head()
```

```python
y=bike_df['price']
```

```python
y.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
X_train.head()
```

```python
ohe=OneHotEncoder()
ohe.fit(X[['bike_name','city','brand']])
```

#### column transformer

one hot encoding -> bike_name, city, brand ordinal encoding -> owner

```python
step1= make_column_transformer((OneHotEncoder(categories=ohe.categories_,sparse=False,drop="first"),[0,1,5])
                              ,remainder='passthrough'
)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
step2=LinearRegression()
```

```python
pipe=make_pipeline(step1,step2)
```

```python
pipe.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
y_pred=pipe.predict(X_test)
print('r2score',r2_score(y_test,y_pred))
print('maescore',mean_absolute_error(y_test,y_pred))
```

### result

the model has 98% accuracy and has 4071 in price difference to orignal price

```python
pipe.predict(pd.DataFrame([['Yamaha FZ','Faridabad',100000,8,150,'Yamaha']],columns=['bike_name','city','kms_driven','age','power','brand']))
```
